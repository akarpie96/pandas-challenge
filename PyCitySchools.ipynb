```python
# Dependencies and Setup
import pandas as pd


# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python

school_name=school_data_complete["school_name"]
number_students=len(school_data_complete["student_name"])


total_schools=len(school_name.unique())



total_budget=sum(school_data["budget"].unique())





math_score_mean=round(school_data_complete["math_score"].mean(),5)



reading_score_mean=round(school_data_complete["reading_score"].mean(),5)



school_data_complete
#Percentage of students who passed math 
passing_score_df=school_data_complete.loc[school_data_complete["math_score"]>=70 , :]
passing_score_math=passing_score_df["math_score"].count()*100
passing_percentage_math=passing_score_math/number_students


#Percentage of students who passed reading
passing_score_df=school_data_complete.loc[school_data_complete["reading_score"]>=70 , :]
passing_score_reading=passing_score_df["reading_score"].count()*100
passing_percentage_reading=passing_score_reading/number_students


#Percentage of students who passed both
passing_score_df=school_data_complete.loc[(school_data_complete["reading_score"] >=70) & (school_data_complete["math_score"]>=70), ["student_name", "reading_score", "math_score"]]
passing_student=passing_score_df["student_name"].count()*100
passing_percentage_total=passing_student/number_students


#Create Data Frame to summarize information
School_DataSummary_df=pd.DataFrame({"Total Schools": [total_schools], 
                                   "Total Students": [number_students],
                                   "Total Budget": [total_budget], 
                                   "Average Math Score": [math_score_mean],
                                   "Average Reading Score": [reading_score_mean], 
                                   "% Passing Math": [passing_percentage_math],
                                   "% Passing Reading": [passing_percentage_reading], 
                                   "% Overall Passing": [passing_percentage_total]})

School_DataSummary_df
```

```python

school_summary_df=school_data_complete[["school_name", "type", "reading_score", "math_score", "size", "budget", "math_score", "reading_score"]]

school_summary_df=school_summary_df.groupby(["school_name"]).mean()

school_summary_df["Budget per student"]=school_summary_df["budget"]/school_summary_df["size"]




passing_score_math_df=school_data_complete.loc[school_data_complete["math_score"]>=70 , :]

passing_score_reading_df=school_data_complete.loc[school_data_complete["reading_score"]>=70 , :]

total_pass_math=passing_score_math_df.groupby(["school_name"]).count()["math_score"]

total_pass_reading=passing_score_reading_df.groupby(["school_name"]).count()["reading_score"]


total_students_school=school_data_complete.groupby(["school_name"]).count()["student_name"]



school_summary_df["% Passing Math"]=(total_pass_math/total_students_school)*100

school_summary_df["% Passing Reading"]=(total_pass_reading/total_students_school)*100

school_summary_df["School Type"]=school_data_complete.groupby(["school_name"])["type"].unique()


passing_both_df=school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70), :]




passing_both_total=passing_both_df.groupby(["school_name"])["math_score"].count()

school_summary_df["% Overall Passing"]=(passing_both_total/total_students_school)*100


school_summary_df=school_summary_df[["School Type",
                                     "size",
                                     "budget", 
                                     "Budget per student", 
                                     "math_score",
                                     "reading_score", 
                                     "% Passing Math",
                                     "% Passing Reading", 
                                     "% Overall Passing"]] 





renamed_Df=school_summary_df.rename(columns={"School Type": "School Type",
                                             "size":"Total Students",
                                             "budget":"Total School Budget",
                                            "math_score": "Average Math Score",
                                            "reading_score":"Average Reading Score"})


school_summary_df


renamed_Df = renamed_Df.loc[:,~renamed_Df.columns.duplicated()]

renamed_Df.index.name=None

renamed_Df

```

```python
Strongest_Schools_Df = renamed_Df.sort_values(["% Overall Passing"], ascending=False)
Strongest_Schools_Df.head()
```

```python
Weakest_Schools_Df=renamed_Df.sort_values(["% Overall Passing"], ascending=True)

Weakest_Schools_Df.head()
```

```python
#Filtering initital data set by grade and creating new DF's associated with respective grade
Ninth_grade_df=school_data_complete.loc[school_data_complete["grade"]=="9th" , :]
Tenth_grade_df=school_data_complete.loc[school_data_complete["grade"]=="10th" , :]
Eleventh_grade_df=school_data_complete.loc[school_data_complete["grade"]=="11th" , :]
Twelfth_grade_df=school_data_complete.loc[school_data_complete["grade"]=="12th" , :]


#Finding Mean Score per grade
Ninth_Grade_Math=Ninth_grade_df.groupby(["school_name"])["math_score"].mean()
Tenth_Grade_Math=Tenth_grade_df.groupby(["school_name"])["math_score"].mean()
Eleventh_Grade_Math=Eleventh_grade_df.groupby(["school_name"])["math_score"].mean()
Twelfth_Grade_Math=Twelfth_grade_df.groupby(["school_name"])["math_score"].mean()

#Summary DF 

Grade_Summary_df=school_data_complete.groupby(["school_name"]).mean()

Grade_Summary_df["9th Grade"]=Ninth_Grade_Math
Grade_Summary_df["10th Grade"]=Tenth_Grade_Math
Grade_Summary_df["11th Grade"]=Eleventh_Grade_Math
Grade_Summary_df["12th Grade"]=Twelfth_Grade_Math


Math_Grade_Summary_df=Grade_Summary_df[["9th Grade", "10th Grade", "11th Grade", "12th Grade"]]

Math_Grade_Summary_df

Math_Grade_Summary_df.index.name=None 

Math_Grade_Summary_df
```

```python
Ninth_Grade_Reading=Ninth_grade_df.groupby(["school_name"])["reading_score"].mean()
Tenth_Grade_Reading=Tenth_grade_df.groupby(["school_name"])["reading_score"].mean()
Eleventh_Grade_Reading=Eleventh_grade_df.groupby(["school_name"])["reading_score"].mean()
Twelfth_Grade_Reading=Twelfth_grade_df.groupby(["school_name"])["reading_score"].mean()

Grade_Reading_Summary_df=school_data_complete.groupby(["school_name"]).mean()

Grade_Reading_Summary_df["9th Grade"]=Ninth_Grade_Reading
Grade_Reading_Summary_df["10th Grade"]=Tenth_Grade_Reading
Grade_Reading_Summary_df["11th Grade"]=Eleventh_Grade_Reading
Grade_Reading_Summary_df["12th Grade"]=Twelfth_Grade_Reading


Grade_Reading_Summary_df=Grade_Reading_Summary_df[["9th Grade", "10th Grade", "11th Grade", "12th Grade"]]

Grade_Reading_Summary_df

Grade_Reading_Summary_df.index.name=None 

Grade_Reading_Summary_df


```

```python
### Filtered by School Type 
#Create Summary Data Frame by 'type'
type_df=school_data_complete.groupby(["type"]).mean()

#Filter initial data set by passed reading
type_reading_df=school_data_complete.loc[school_data_complete["reading_score"]>=70,:]
#Groupby 'type' in filtered data and pass in reading_score.count to find total number of passed reading by type. 
#Divide above by total number of students 
type_passed_reading=(type_reading_df.groupby("type")["reading_score"].count()/school_data_complete.groupby(["type"])["reading_score"].count())*100

#Filter initial data to find all students who passed math
type_math_df=school_data_complete.loc[school_data_complete["math_score"]>=70,:]

#Groupby 'type' in filtered data and pass in math_score.count to find total number of students who passed math by type. 
#Divide above by total number of students, assigning new series to type_passed_math
type_passed_math=(type_math_df.groupby("type")["math_score"].count()/school_data_complete.groupby(["type"])["math_score"].count())*100

type_passed_math

#Filter initial data set to find students who passed both

type_passed_both_df=school_data_complete.loc[(school_data_complete["math_score"]>=70) & (school_data_complete["reading_score"]>=70), :]

#Groupby 'type' in filtered data and pass in math_score.count or reading_score.count to find total number of students who passed both 
#Divide above by total number of students, assigning new series to type_passed_both

type_passed_both=(type_passed_both_df.groupby("type")["math_score"].count()/school_data_complete.groupby(["type"])["math_score"].count())*100

type_passed_both

#Add new series/columns to summary data set 

type_df["% Passing Math"]=type_passed_math

type_df["% Passing Reading"]=type_passed_reading

type_df["% Overall Passing"]=type_passed_both

type_df=type_df[["reading_score", "math_score", "% Passing Math", "% Passing Reading", "% Overall Passing"]]

type_df



```
